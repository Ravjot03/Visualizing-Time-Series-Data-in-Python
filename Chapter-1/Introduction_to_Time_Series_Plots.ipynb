## Data Visualizations

### Basic time series plot

Plotting the time series

```
df.plot()
plt.show()  #show plot
```

### Adding labels to the plot

Specifying the x-axis label and y-axis label to the plot.

```
ax = df.plot(color='red')  #setting the color of the line to red
ax.set_xlabel('Date')  #x-axis label
ax.set_ylabel('Number of Great Discoveries')  #y-axis label
plt.show() #show plot
```

### Changing the style of the plot

The matplotlib library also comes with a number of built-in stylesheets that allows to customize the appearance of our plots.

```
plt.style.use('ggplot')  #setting the plot style as "ggplot"
ax1 = df.plot(color='green')  #setting the color of the line to green
ax1.set_xlabel('Date')  #x-axis label
ax1.set_ylabel('Number of Great Discoveries')  #y-axis label
plt.show()  #show plot
```

```
plt.style.use('fivethirtyeight')  #setting the style of plot to "fivethirtyeight" style
ax2 = df.plot(color='blue')  #setting the color of the line as blue
ax2.set_xlabel('Date') #x-axis label
ax2.set_ylabel('Number of Great Discoveries')  #y-axis label
plt.show()  #show plot
```

### Formatting the plot style -
1. Changing the dimensions of the plot
2. Changing the width of the line in the plot
3. Changing the color of the line
4. Specifying font size for the title of the plot

```
# Setting the color of the line as blue, dimensions of the plot as (8, 4), width of the line in plot equals to 2, and fontsize equals to 6 
ax = df.plot(color='blue', figsize=(8, 4), linewidth=2, fontsize=6)

# Specify the title in the plot
ax.set_title('Number of great inventions and scientific discoveries from 1860 to 1959', fontsize=8)

plt.show()  #show plot
```

Changing the plot style back to default

```
plt.style.use('default')
```

### Subsetting the time series

When plotting time series data, we may occasionally want to visualize only a subset of the data. The pandas library provides powerful indexing and subsetting methods that allow us to extract specific portions of a DataFrame.

Subsetting the time series data using the index, and plotting the data

```
df1 = df['1945-01-01':'1950-01-01']  #making a subset that contains all the data between January 1, 1945 and January 1, 1950
df1.plot()  #plotting the time series
plt.show()  #show plot
```

```
df2 = df[:'1900-12-31']  #making a subset that contains all data till December 31, 1900
df2.plot() #plotting the data
plt.show()  #show plot
```

### Adding Vertical & Horizontal Markers

Additional annotations can help further emphasize specific observations or events.

By adding markers at specific timestamps of our time series plot, we can highlight significant events.

#### Vertical Markers

`.axvline()` - This function adds the vertical lines across the axes of the plot.



```
ax = df.plot()  #plotting the data
ax.axvline('1929-01-01', color='red')  #adding red coloured vertical line at a timestamp of 1929-01-01
plt.show() #show plot
```

#### Horizontal Markers

`.axhline()` - It is used to add a horizontal line across the axis.

```
ax = df.plot()  #plotting the time series data
ax.axhline(4, color='green', linestyle='--')  #adding green coloured dashed line at 4
plt.show()  #show plot
```

```
ax = df.plot()  #plotting the time series data
ax.axhline(df['Y'].mean(), color='green', label='Average Y')  #adding horizontal line which shows the Average Y
plt.legend() #showing the labels
plt.show()  #show plot
```

#### Both Vertical & Horizontal Markers

```
ax = df.plot()  #plotting the time series data
ax.axhline(5, color='brown', linestyle='--')  #adding horizontal line to the plot
ax.axvline('1900-01-01',color='orange',linestyle='-.')  #adding vertical line to the plot
plt.show()  #show plot
```

### Adding Shaded Regions

When plotting time series data in Python, it is also possible to highlight complete regions of the time series plot.

#### Vertically Shading Regions

`.axvspan()` - This function sets the vertical rectangle across the axes of the plot.

```
ax = df.plot()  #plotting the time series data
ax.axvspan('1895-01-01', '1915-01-01', color='red', alpha=0.3)  #adding red coloured vertical shaded region b/w 1895-01-01 and 1915-01-01
plt.show()  #show plot
```

#### Horizontally Shading Regions

`.axhspan()` - This function sets the horizontal rectangle across the axes of the plot.

```
ax = df.plot()  #plotting the time series data
ax.axhspan(3, 5, color='green', alpha=0.3)  #adding green coloured horizontal shaded region b/w points 3 and 5
plt.show()  #show plot
```

#### Both Vertically & Horizontally shading regions

```
ax = df.plot()  #plotting the time series data
ax.axvspan('1890-01-01','1900-01-01', color='orange', alpha=0.3)  #adding orange coloured vertical shaded region
ax.axhspan(4, 6, color='brown', alpha=0.3)  #adding brown coloured horizontal shaded region
plt.show()  #show plot
```

### Markers & Shading Regions

```
ax = df.plot()  #plotting the time series data
ax.axhline(5, color='brown', linestyle='--')  #adding horizontal line/marker
ax.axvline('1900-01-01',color='orange',linestyle='--')  #adding vertical line/ marker
ax.axvspan('1895-01-01','1905-01-01', color='orange', alpha=0.3)  #adding vertical shaded region
ax.axhspan(4, 6, color='brown', alpha=0.3)  #adding horizontal shaded region
plt.show()  #show plot
```
